```python
# text data
data = open("../exploratory/data/input.txt", 'r').read()
```

```python
len(data) / 30
```

```python
# setup params
sequence_length = 25
vocab_size = len(set(data))
```

### Define LSTM trainer 

* Preprocesses text data, wraps around upon reaching end of file etc.

```python
# define network
lstm_nn = SequentialNeuralNetwork(
    num_features = vocab_size,
    sequence_length = sequence_length,
    layers = [LSTMLayer(neurons=250)],
    loss = LogSoftmaxLoss())
```

```python
# define optimizer
optimizer = SGD(0.01)
```

```python
# define trainer
trainer = LSTMTrainer(lstm_nn, optimizer, data)
```

```python
# API for training
trainer.fit(epochs = 10,
       eval_every = 1,
       batch_size=25,
       seed=102618,
       single_output=True);
```
